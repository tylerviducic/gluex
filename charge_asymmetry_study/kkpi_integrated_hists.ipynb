```python
import ROOT
import my_library.common_analysis_tools as ct
import my_library.kinematic_cuts as kcuts
import my_library.constants as constants
```

```python
ROOT.gStyle.SetOptStat(0)
ROOT.EnableImplicitMT()
```

```python
df_pipkmks = ct.get_dataframe('pipkmks', 'gluex1', 'data')
df_pimkpks = ct.get_dataframe('pimkpks', 'gluex1', 'data')
```

```python
hist_pipkmks = df_pipkmks.Histo1D(('pipkmks', 'M(#pi^{+}K^{-}K_{S})', 100, 1.1, 2.1), 'pipkmks_m')
hist_pimkpks = df_pimkpks.Histo1D(('pimkpks', 'M(#pi^{-}K^{+}K_{S})', 100, 1.1, 2.1), 'pimkpks_m')
```

```python
hist_pipkmks.SetLineColor(ROOT.kBlack)
hist_pimkpks.SetLineColor(ROOT.kBlack)

c = ROOT.TCanvas('c', 'c', 500, 1000)
c.Divide(1,2)
c.cd(1)
hist_pimkpks.Draw()
c.cd(2)
hist_pipkmks.Draw()
c.Draw()
```

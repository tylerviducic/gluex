```python
import ROOT
import my_library.common_analysis_tools as tools
import my_library.constants as constants
import my_library.kinematic_cuts as kcuts
```

```python
ROOT.EnableImplicitMT()
ROOT.gStyle.SetOptStat(0)
```

```python
pipkmks_df = tools.get_dataframe('pipkmks', 'spring', 'data', filtered=False)
pimkpks_df = tools.get_dataframe('pimkpks', 'spring', 'data', filtered=False)
```

```python
pipkmks_df = pipkmks_df.Filter(kcuts.KINFIT_CL_CUT).Filter(kcuts.MX2_PPIPKMKS_CUT).Filter(kcuts.KS_PATHLENGTH_CUT).Filter(kcuts.KMP_MASS_CUT).Filter(kcuts.P_P_CUT)
pimkpks_df = pimkpks_df.Filter(kcuts.KINFIT_CL_CUT).Filter(kcuts.MX2_PPIMKPKS_CUT).Filter(kcuts.KS_PATHLENGTH_CUT).Filter(kcuts.KSP_MASS_CUT).Filter(kcuts.P_P_CUT)
```

```python
hist_ppip = pipkmks_df.Histo1D(('ppim', 'M(p#pi^{+})', 200, 1.0, 1.3), 'ppip_m')
hist_ppim = pimkpks_df.Histo1D(('ppim', 'M(p#pi^{-})', 200, 1.0, 1.3), 'ppim_m')
```

```python
c = ROOT.TCanvas('c', 'c', 800, 600)
c.Divide(2,1)
c.cd(1)
hist_ppip.Draw()
c.cd(2)
hist_ppim.Draw()
c.Draw()
```

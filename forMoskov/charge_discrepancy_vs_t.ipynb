```python
import ROOT
import my_library.common_analysis_tools as tools
import my_library.constants as constants
import my_library.kinematic_cuts as kcuts
```

```python
ROOT.EnableImplicitMT()
ROOT.gStyle.SetOptStat(0)
```

```python
df_pipkmks = tools.get_dataframe('pipkmks', 'spring', 'data')
df_pimkpks = tools.get_dataframe('pimkpks', 'spring', 'data')
```

```python
df_pipkmks = df_pipkmks.Filter(kcuts.KSTAR_ALL_CUT_PIPKMKS)
df_pimkpks = df_pimkpks.Filter(kcuts.KSTAR_ALL_CUT_PIMKPKS)
```

```python
hists_pipkmks = []
hists_pimkpks = []

for e in range(8, 12):
    cut = f'e_beam < {e+0.5} && e_beam > {e-0.5}'
    print(cut)
    hist_pipkmks = df_pipkmks.Filter(cut).Histo1D((f'pipkmks_{e}', 'M(#pi^{+}K^{-}K_{s}) for E_{#gamma} = ' + str(e), 60, 1.1, 1.7), 'pipkmks_m')
    hist_pimkpks = df_pimkpks.Filter(cut).Histo1D((f'pimkpks_{e}', 'M(#pi^{-}K^{+}K_{s}) for E_{#gamma} = ' + str(e), 60, 1.1, 1.7), 'pimkpks_m')
    hists_pipkmks.append(hist_pipkmks)
    hists_pimkpks.append(hist_pimkpks)
```

```python
c = ROOT.TCanvas('c', 'c', 800, 600)
c.Divide(2, 2)
for i, hist_pip in enumerate(hists_pipkmks):
    c.cd(i+1)
    hist_pip.SetLineColor(ROOT.kBlue)
    hists_pimkpks[i].SetLineColor(ROOT.kRed)
    hists_pimkpks[i].Draw()
    hist_pip.Draw('same')
    c.Update()
c.Draw()
```

```python
from my_library.constants import T_CUT_DICT


hists_pipkmks_8 = []
hists_pimkpks_8 = []
df_pipkmks_8 = df_pipkmks.Filter('e_beam < 8.5 && e_beam > 7.5')
df_pimkpks_8 = df_pimkpks.Filter('e_beam < 8.5 && e_beam > 7.5')
for cut in constants.T_CUT_DICT:
    hists_pipkmks_8.append(df_pipkmks_8.Filter(f'mand_t > {T_CUT_DICT[cut][0]} && mand_t < {T_CUT_DICT[cut][1]}').Histo1D((f'pipkmks_{T_CUT_DICT[cut]}', 'M(#pi^{+}K^{-}K_{s}) for t = ' + (str(T_CUT_DICT[cut][1] - (T_CUT_DICT[cut][1] - T_CUT_DICT[cut][0]))), 60, 1.1, 1.7), 'pipkmks_m'))
    hists_pimkpks_8.append(df_pimkpks_8.Filter(f'mand_t > {T_CUT_DICT[cut][0]} && mand_t < {T_CUT_DICT[cut][1]}').Histo1D((f'pimkpks_{T_CUT_DICT[cut]}', 'M(#pi^{-}K^{+}K_{s}) for t = ' + (str(T_CUT_DICT[cut][1] - (T_CUT_DICT[cut][1] - T_CUT_DICT[cut][0]))), 60, 1.1, 1.7), 'pimkpks_m'))
```

```python
c1 = ROOT.TCanvas('c1', 'c1', 800, 600)
c1.Divide(4, 2)
for i, hist_pip in enumerate(hists_pipkmks_8):
    c1.cd(i+1)
    hist_pip.SetLineColor(ROOT.kBlue)
    hists_pimkpks_8[i].SetLineColor(ROOT.kRed)
    hists_pimkpks_8[i].Draw()
    hist_pip.Draw('same')
    c1.Update()
c1.Draw()
```

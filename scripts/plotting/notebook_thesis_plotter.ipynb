```python
import ROOT
```

```python
data_file = '/home/viducic/thesis_work/data/pipkmks/data/bestX2/pipkmks_flat_bestX2_2017.root'
tree_name = 'pipkmks__B4_M16'
```

```python
df = ROOT.RDataFrame(tree_name, data_file)

df = df.Define('ks_px', 'pip2_px + pim_px')
df = df.Define('ks_py', 'pip2_py + pim_py')
df = df.Define('ks_pz', 'pip2_pz + pim_pz')
df = df.Define('ks_E', 'pip2_E + pim_E')
df = df.Define('ks_m', 'sqrt(ks_E*ks_E - ks_px*ks_px - ks_py*ks_py - ks_pz*ks_pz)')

hist = df.Histo1D(('ksm', 'ksm', 500, 0.2, 0.8), 'ks_m')
```

```python
%matplotlib inline
c1 = ROOT.TCanvas()
hist.Draw()
c1.Draw()
c1.Update()
```

```python
import ROOT
import my_library.common_analysis_tools as tools
import my_library.constants as constants 
import my_library.gluex_style 
```

```python
ROOT.EnableImplicitMT()
```

```python
pipkmks_thrown_df = tools.get_dataframe('pipkmks', 'fall', 'signal', filtered=False, thrown=True)
```

```python
print(pipkmks_thrown_df.GetColumnNames())
```

```python
hist_kkpi = pipkmks_thrown_df.Histo1D(('hist_kkpi', 'M(K^{-}K_{s}#pi^{+}) [GeV]', 200, 1.1, 1.5), 'pipkmks_m')
hist_ks = pipkmks_thrown_df.Histo1D(('hist_ks', ';M(#pi^{+}#pi^{-}) [GeV]', 200, 0.4, 0.6), 'mpippim')
```

```python
c = ROOT.TCanvas('c', 'c', 800, 600)
c.Clear()

hist_kkpi.Draw()
hist_kkpi.GetXaxis().SetTitle('Generated M(K^{-}K_{s}#pi^{+}) [GeV]')
hist_kkpi.GetYaxis().SetTitle('Counts/2 MeV')
c.Draw()
```

```python
c.Clear()

hist_ks.Draw()
hist_ks.GetXaxis().SetTitle('Generated M(#pi^{+}#pi^{-}) [GeV]')
hist_ks.GetYaxis().SetTitle('Counts/ MeV')
c.Draw()
```

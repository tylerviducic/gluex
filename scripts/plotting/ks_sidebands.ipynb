```python
ROOT.EnableImplicitMT()
```

```python
channel = 'pipkmks'
run_period = 'spring'
data_type = 'data'
```

```python
c = ROOT.TCanvas()
```

```python
df = ct.get_dataframe(channel, run_period, data_type, filtered=False)
```

```python
df.Filter(ct.MX2_PPIPKMKS_CUT).Filter(ct.KS_PATHLENGTH_CUT).Filter(ct.PPIP_MASS_CUT).Filter(ct.KMP_MASS_CUT).Filter(ct.P_P_CUT)
```

```python
hist_ks_vs_kkp = df.Histo2D(('hist_ks_vs_kkp', 'M(K_{s}) vs M(KK#pi)', 30, 1.2, 1.5, 100, 0.3, 0.7), f'{channel}_m', 'ks_m')
```

```python
hist_ks_vs_kkp.Draw('colz')
c.Draw()
```

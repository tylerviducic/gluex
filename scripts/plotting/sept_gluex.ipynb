```python
import ROOT
import my_library.common_analysis_tools as tools
import my_library.constants as constants
import my_library.kinematic_cuts as kcuts
```

```python
ROOT.EnableImplicitMT()
```

```python
df_pipkmks = tools.get_dataframe('pipkmks', 'spring', 'data', filtered=False)
df_pimkpks = tools.get_dataframe('pimkpks', 'spring', 'data', filtered=False)
```

```python
df_pipkmks = df_pipkmks.Filter('chi2ndf < 4').Filter(kcuts.P_P_CUT).Filter(kcuts.KS_MASS_CUT)\
    .Filter(kcuts.KS_PATHLENGTH_CUT).Filter(kcuts.MX2_PPIPKMKS_CUT)
df_pimkpks = df_pimkpks.Filter('chi2ndf < 4').Filter(kcuts.P_P_CUT).Filter(kcuts.KS_MASS_CUT)\
    .Filter(kcuts.KS_PATHLENGTH_CUT).Filter(kcuts.MX2_PPIMKPKS_CUT)
```

```python
hist_ppip = df_pipkmks.Histo1D(('ppip', 'M(p#pi^{+})', 400, 1.0, 2.0), 'ppip_m')
hist_ppip.GetXaxis().SetTitle('M(p#pi^{+}) [GeV/c^{2}]')
hist_ppip.GetYaxis().SetTitle('Counts')
hist_ppip.SetLineColor(ROOT.TColor.GetColor(constants.COLORBLIND_HEX_DICT['blue']))

hist_ppim = df_pimkpks.Histo1D(('ppim', 'M(p#pi^{-})', 400, 1.0, 2.0), 'ppim_m')
hist_ppim.GetXaxis().SetTitle('M(p#pi^{-}) [GeV/c^{2}]')
hist_ppim.GetYaxis().SetTitle('Counts')
hist_ppim.SetLineColor(ROOT.TColor.GetColor(constants.COLORBLIND_HEX_DICT['red']))

```

```python
ROOT.gStyle.SetOptStat(0)
```

```python
c = ROOT.TCanvas('c', 'c', 800, 600)
hist_ppip.Draw()
c.Draw()
```

```python
hist_ppim.Draw()
c.Draw()
```

```python
df_pipkmks = df_pipkmks.Filter(kcuts.PPIP_MASS_CUT)
df_pimkpks = df_pimkpks.Filter(kcuts.PPIM_MASS_CUT)
```

```python
hist_kmp = df_pipkmks.Histo1D(('kmp', 'M(K^{-}p)', 400, 1.0, 3.0), 'kmp_m')
hist_kmp.GetXaxis().SetTitle('M(K^{-}p) [GeV/c^{2}]')
hist_kmp.GetYaxis().SetTitle('Counts')
hist_kmp.SetLineColor(ROOT.TColor.GetColor(constants.COLORBLIND_HEX_DICT['blue']))
hist_ksp = df_pimkpks.Histo1D(('kpp', 'M(K^{+}p)', 400, 1.0, 3.0), 'ksp_m')
hist_ksp.GetXaxis().SetTitle('M(K^{+}p) [GeV/c^{2}]')
hist_ksp.GetYaxis().SetTitle('Counts')
hist_ksp.SetLineColor(ROOT.TColor.GetColor(constants.COLORBLIND_HEX_DICT['red']))
```

```python
hist_kmp.Draw()
c.Draw()
```

```python
hist_ksp.Draw()
c.Draw()
```

# Plot and Fit The $ K^{*} $ cut efficiency vs $ M(KK  \pi ) $

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from my_library.common_analysis_tools import COLORBLIND_HEX_DICT
```

#### Load in the 10MeV and 1MeV csv files to dataframes

```python
df_1 = pd.read_csv('/work/halld/home/viducic/data/ps_dalitz/kstar_cut_efficiency_1.0.csv')
df_10 = pd.read_csv('/work/halld/home/viducic/data/ps_dalitz/kstar_cut_efficiency_stepsize_10.csv')
```

#### Plot both datasets next to each other

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))
fig.suptitle("$K^{*}$ Rejection Efficiency vs. $M(KK\\pi)$", fontsize=16)
ax1.scatter(df_10.mass_bin_center, df_10.kstar_cut_efficiency, s=15, color=COLORBLIND_HEX_DICT['blue'])
ax1.set_title("step size 10 MeV")
ax1.set_xlabel("$M(KK\\pi)$ [GeV]")
ax1.set_ylabel("$K^{*}$ Rejection Efficiency")
ax2.scatter(df_1.mass_bin_center, df_1.kstar_cut_efficiency, s=5, color=COLORBLIND_HEX_DICT['red'])
ax2.set_title("step size 1 MeV")
ax2.set_xlabel("$M(KK\\pi)$ [GeV]")
ax2.set_ylabel("$K^{*}$ Rejection Efficiency")
```
